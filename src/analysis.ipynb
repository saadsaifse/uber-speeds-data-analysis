Load required python libraries

```python
import numpy as np
import pandas as pd
import os
```

```python
dataPath = os.path.join("..", "data", "london2018")
januarySpeedsData = pd.read_csv(os.path.join(dataPath, "movement-speeds-hourly-london-2018-1.csv"))

januarySpeedsData
```

**Task:** For each OSM Way Id, sort the rows with ascending order of days and hours.

```python
januarySpeedsData = januarySpeedsData.sort_values(by=['osm_way_id', 'day', 'hour'])
januarySpeedsData
```

**Task:** For each OSM Way Id, find the overall maximum, minimum and mean speeds.

```python
januarySpeedsDataGrouped = januarySpeedsData.groupby(['osm_way_id'])
```

**Task:** What are the max, min and mean speeds for 

```python
speed_stats = pd.DataFrame({
    'min_speed': januarySpeedsDataGrouped['speed_mph_mean'].min(),
    'max_speed': januarySpeedsDataGrouped['speed_mph_mean'].max(),
    'mean_speed': januarySpeedsDataGrouped['speed_mph_mean'].mean()
    })
speed_stats
```

**Task:** Get mean speed per day as hourly speeds data will be too detailed and hard to deal with. Get the daily time series of the first street (for now).

```python
firstWay = januarySpeedsDataGrouped.get_group(73)
firstWayByDay = firstWay.groupby(['day'])
streetSpeedData = firstWayByDay['speed_mph_mean'].mean()
```

**Task:** Plot the time series

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('Day')
ax.set_ylabel('Mean Speed')
ax.plot(streetSpeedData)
```

**Task:** What is the percentage change in the daily speeds?

```python
speedPercentChange = streetSpeedData.pct_change()
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('Day')
ax.set_ylabel('Percent change in speed')
ax.plot(speedPercentChange)
```

If we look at the above street data, there exists two speed observations for `2018-01-01 00:00:00`. This is due to the fact that this is a two-way road and that is why for these two observations, `osm_start_node_id` and `osm_end_node_id` have identical but opposite values.
To make our time series and calculations simpler, we can only take observaions for only a single way. To do that, lets group by `osm_start_node_id` and `osm_end_node_id` as well. 

```python
secondStreet = secondStreet.groupby(['osm_way_id', 'osm_start_node_id', 'osm_end_node_id'])
```
