### Task: Desribe the dataset

The datasets contain hourly speed observations of three months. In Open Street Map (OSM), a way defines a line connecting two nodes e.g., a street or a road. Whereas OSM nodes are the points in real world such as a junction. For the speed analysis, a street can be uniquely identified via `osm_way_id`, `osm_start_node_id`, and `osm_end_node_id`. Note that we need all three parameters to identify a street speed because a street can be one way or two way. In case of later, `osm-way-id` will be the same for both ways but their node ids would swap.

Mean Speeds in miles per hour are provided per hour. But in the dataset, their are missing values and data is not available for each hour.

### Task: Visualize autocorrelation of the time series and analyse the result

Correlation defines the strength of the relationship between two variables. In a single time series, we can calculate the correlation between current and previous observations (called lags) of the same time series. This is also called autocorrelation.

Lets find and visualize the autocorrelation for Oxford Street speed time series.

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(oxfordSegmentTimeSeries)
```

The above plot shows correlation values (also called correlation coefficients) for different lag values of our time series. As the values are close to 0, it shows weak relationships between the speed values. On the other hand 1 or -1 show strong correlations.

We can also find autocorrelation using `statsmodel` library, which gives us different kind of graph.

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(oxfordSegmentTimeSeries)
plt.show()
```

Both plots show Lag-1 autocorrelation to be less than 0.1, that shows really weak correlation between speed values on this road segment.

### Task: Visualize partial autocorrelation of the time series and analyse the result.

Partial autocorrelation removes the intervening indirect correlations with observations from the previous time step and only considers the direct correlations with specified lag observations.

### Task: Decompose time series to find patterns such as Trend, Seasonality or Cycles.

With `statsmodels` we will be able to see the trend, seasonal, and residual components of our time series. There are two kinds of models, `additive` and `multiplicative`. We can use an additive model when it seems that the trend is more linear and the seasonality and trend components seem to be constant over time. Which is exactly what we have used below.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(oxfordSegmentTimeSeries['speed_mph_mean'], model='additive', period=int(oxfordSegmentTimeSeries.size/4))
result.plot()
```

In the above plot, Trend is linear in for the first two months and then declines slightly in March. While seasonal values exist in the range of -1 to 5 and is quite unpredictable due to a lot of variations.

### Task: Perform time series forecasting using ARIMA model

ARIMA stands for AutoRegressive Integrated Moving Average. It is a class of statistical models that capture different standard temporal structures in a time series. 
We have tried two differnet libraries to best fit the ARIMA model and do forecasting.

1. Using `statsmodel` without `auto.arima`. Here we specified the values of p, d, and q parameters. 

We have used `ARIMA(2,1,2) which sets the lag value to 2 for autoregression, uses a difference order of 1 to make the time series stationary, and uses a moving average model of 2.

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
model = ARIMA(oxfordSegmentTimeSeries, order=(2, 1, 2))
modelFit = model.fit()
modelFit.summary()
```

```python
# plot residual errors
residuals = pd.DataFrame(modelFit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

Below code forms a Rolling Forecast ARIMA model that recreates the model after new observations. First it splits the values into training and test data, which for faster execution is set to 95% of training data and 5% of test data. This can also be done 66% and 33% respectively but that takes quiet a lot of time to execute.

Then it fits the model on new history which is a combination of training data and test data and fits the model again in the subsequent loop.

```python
speeds = oxfordSegmentTimeSeries.values
size = int(len(speeds) * 0.95)
train, test = speeds[0:size], speeds[size:len(speeds)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
	model = ARIMA(history, order=(2,1,2))
	model_fit = model.fit(disp=0)
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = test[t]
	history.append(obs)	
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```

The above model did not produce very accurate predictions. Lets use `auto-arima` from `pmdarima` library.

```python
import pmdarima as pm
from pmdarima.model_selection import train_test_split

train, test = train_test_split(oxfordSegmentTimeSeries, train_size=1500)

model = pm.auto_arima(train, start_p=1, start_q=1,
                           max_p=3, max_q=3, m=12,
                           start_P=0, seasonal=True,
                           d=1, D=1, trace=True,
                           error_action='ignore',   
                           stepwise=True)

forecasts = model.predict(test.shape[0])
```

```python
x = np.arange(oxfordSegmentTimeSeries.shape[0])
plt.plot(x[:1500], train, c='blue')
plt.plot(x[1500:], forecasts, c='green')
plt.show()
```

The predicions from `auto.arima` are also about the same range as from the above method. It can be concluded that due to the nature of this time series, the prediction values are coming out in the range of 13-14 mph
