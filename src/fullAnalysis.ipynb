### Task: Desribe the dataset

The datasets contain hourly speed observations of three months. In Open Street Map (OSM), a way defines a line connecting two nodes e.g., a street or a road. Whereas OSM nodes are the points in real world such as a junction. For the speed analysis, a street can be uniquely identified via `osm_way_id`, `osm_start_node_id`, and `osm_end_node_id`. Note that we need all three parameters to identify a street speed because a street can be one way or two way. In case of later, `osm-way-id` will be the same for both ways but their node ids would swap.

Mean Speeds in miles per hour are provided per hour. But in the dataset, their are missing values and data is not available for each hour.

### Task: Shortlist a road where time series analysis can be made.

Selecitng a road is a challenging task, as we are unaware about whether the complete data will be available for that road.

Digging into the london 2018 geometries data, there exists OSM Way Ids and road names, along wiht OSM Highway type, among other data.

Here we select the "Park Street Lane" of all highway types.

```python
parkStreetLane = london2018Geometry[(london2018Geometry['osmname'].str.contains('Park street lane', case=False, na=False))]

#& (london2018Geometry['osmhighway'].str.contains('tertiary'))]
parkStreetLane.head()
```

```python
parkStreetWayIds = parkStreetLane['osmwayid'].unique()
parkStreetWayIds
```

```python
parkStreetData = data[data['osm_way_id'].isin(parkStreetWayIds)]
parkStreetData
```

Parklane street has less number of observations for the speed data. Lets try Oxford Street, which is one of the busiest street of the city.

```python
oxfordStreet = london2018Geometry[(london2018Geometry['osmname'].str.match('Oxford street', case=False, na=False))]

#& (london2018Geometry['osmhighway'].str.contains('tertiary'))]
oxfordStreet
```

```python
oxfordStreetWayIds = oxfordStreet['osmwayid'].unique()
oxfordStreetWayIds
```

```python
oxfordSreetData = data[data['osm_way_id'].isin(oxfordStreetWayIds)]
oxfordSreetData
```

Segment Ids and Junction Ids represent OSM Way Id and Node Ids, however, there is a difference. Segments are much smaller from relatively longer Way Ids. That means, there is one to many relationship between OSM Way Id and Segment Ids. Here we will analyse speeds per segment and will map them later to OSM Way Ids if required for plotting, etc.

```python
oxfordStreetGroup = oxfordSreetData.groupby(['segment_id', 'start_junction_id', 'end_junction_id'])
size = oxfordStreetGroup.size()
size
```

Lets take a segment with maximum observations for analysis

```python
maxSize = size.max()
oxfordStreetSegment = None
for grp, data in oxfordStreetGroup:
    if len(data) == maxSize:
        oxfordStreetSegment = data
oxfordStreetSegment
```

Convert `utc_timestamp` from plain objects to DateTime objects

```python
oxfordStreetSegment['utc_timestamp'] = pd.to_datetime(oxfordStreetSegment['utc_timestamp'])
```

Create time series with date time as index and speed values as values

```python
oxfordSegmentTimeSeries = oxfordStreetSegment[['utc_timestamp', 'speed_mph_mean']].set_index('utc_timestamp')
oxfordSegmentTimeSeries
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=oxfordSegmentTimeSeries.index, y=oxfordSegmentTimeSeries['speed_mph_mean']))

fig.update_layout(title_text='Oxford Street Segment Time Series',
                  xaxis_rangeslider_visible=True)
fig.show()
```

Check if daily speed observations exist for each hour in all 3 months.

```python
from pandas import Grouper
groups = oxfordSegmentTimeSeries.groupby(Grouper(freq='D'))
for key, group in groups:
    print(f'Day {key.day} has {group.size} observations')
```

Only few observations are missing, but we can definitely live without them ;)

### Task: Find out how the speeds are disrubuted without temporal ordering i.e., how the actual speed values appeared in the time series.

```python
import plotly.express as px
fig = px.histogram(oxfordSegmentTimeSeries, x=oxfordSegmentTimeSeries['speed_mph_mean'], y=oxfordSegmentTimeSeries['speed_mph_mean'])
fig.show()
```

The above historgram has the actual speed values on the `x-axis` and the number of times (count) of their appearance in the time series dataset. This shows that the speed of about `14mph` was observed the most. The bins in the plot are automatically selected by the plot function according to the spread of data.
