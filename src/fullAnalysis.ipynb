### Task: Desribe the dataset

The datasets contain hourly speed observations of three months. In Open Street Map (OSM), a way defines a line connecting two nodes e.g., a street or a road. Whereas OSM nodes are the points in real world such as a junction. For the speed analysis, a street can be uniquely identified via `osm_way_id`, `osm_start_node_id`, and `osm_end_node_id`. Note that we need all three parameters to identify a street speed because a street can be one way or two way. In case of later, `osm-way-id` will be the same for both ways but their node ids would swap.

Mean Speeds in miles per hour are provided per hour. But in the dataset, their are missing values and data is not available for each hour.

### Task: Shortlist Roads

The above dataset has atleast as many time series as the number of roads. We can ofcourse run our analysis on each on of them. But for now, we need to take only a few roads. The decision of which road to take can be derived from collisions dataset. We can analyse collision dataset and take the roads where there were most number of collisions during this time interval.

```python
collisionData = pd.read_csv(os.path.join(dataPath, 'dftRoadSafetyData_Accidents_2018.csv'))
collisionData
```

```python
import geopandas
gdf = geopandas.GeoDataFrame(
    collisionData, geometry=geopandas.points_from_xy(collisionData.Longitude, collisionData.Latitude))
gdf
```

```python
collisionData.Date = pd.to_datetime(collisionData.Date)
```

Get 3 months collision data only to relate with our 3 months speed dataset

```python
collisions = collisionData[(collisionData.Date >= '2018-01-01') & (collisionData.Date <= '2018-03-31')]
collisions
```

```python
collisionsGroup = collisions.groupby(['Date'])
```

```python
c = collisionsGroup.count()
date = c['Accident_Index'].idxmax()
print(f"Maximum number of accidents ({c.max()['Accident_Index']}) occured on {date.strftime('%Y-%m-%d')}")
```

```python
dayWithMaxAccidents = collisionsGroup.get_group(date)
dayWithMaxAccidents
```
